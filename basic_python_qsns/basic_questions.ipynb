1.Take 5 numbers → print mean, min, max

```python
number=[]
total=0
for i in range(5):
    num=int(input(f"enter the number {i+1}"))
    number.append(num)
for num in number:
    total=total+num
min=number[0]
max=number[0]
for num in number:
    if num<min:
        min=num 
    if num>max:
        max=num 

print(f"the avg is {total/5}\n the minimum num is {min} \n max num is {max}")
```

2. Convert Celsius to Fahrenheit

```python
#  formula to convert the temp from celsius to fahrenheit is 
# F=(C*9/5)+32

c=float(input("enter the temp in celsius : "))
F=(c*9/5)+32

print(f"the temp in farenheit is {F}")
```

3. Input salary → cut 10% tax → print final salary

```python
salary=float(input("enter the salary : "))
final_salary=salary-salary/10
print(f"the final salary after cutdown is {final_salary}")
```

4. Input 3 subject marks → percentage + pass/fail

```python
n=int(input("enter the number subjects : "))
mark=[]
for i in range(n):
    marks=float(input("Enter the marks of subjects "))
    mark.append(marks)
total=sum(mark)
percentage=(total/(n*100))*100
print(percentage)
if(percentage>40):
    print("pass")
else:print("fail")
```
